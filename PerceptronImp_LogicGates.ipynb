```python
inputs=[(0,0),(0,1),(1,0),(1,1)]
#AND
for x1,x2 in inputs:
    s=x1*1+x2*1
    if s>=1.5:
        y=1
    else:
        y=0
    print(x1,x2," ",y)
```

```python
#OR
for x1,x2 in inputs:
    s=x1*1+x2*1
    if s>=0.5:
        y=1
    else:
        y=0
    print(x1,x2," ",y)
```

```python
#NAND
for x1,x2 in inputs:
    s=x1*(-1)+x2*(-1)
    if s>=-1.5:
        y=1
    else:
        y=0
    print(x1,x2," ",y)
```

```python
#NOR
for x1,x2 in inputs:
    s=x1*(-1)+x2*(-1)
    if s>=-0.5:
        y=1
    else:
        y=0
    print(x1,x2," ",y)
```
